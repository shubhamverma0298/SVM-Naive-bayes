#**Theoretical**

**Q 1 What is a Support Vector Machine (SVM)?**
  -A Support Vector Machine (SVM) is a powerful supervised machine learning algorithm used for both classification and regression tasks

**Q 2 What is the difference between Hard Margin and Soft Margin SVM?**
  - In Support Vector Machines (SVMs), hard margin SVM aims to find a hyperplane that perfectly separates data points into different classes with no errors, while soft margin SVM allows for some misclassifications to better handle data that isn't perfectly separable.

**Q 3 What is the mathematical intuition behind SVM?**
  - The core mathematical intuition behind Support Vector Machines (SVMs) is to find the optimal hyperplane that best separates data points into different classes, while also maximizing the margin (distance) between the hyperplane and the closest data points of each class

**Q 4 What is the role of Lagrange Multipliers in SVM?**
  - In Support Vector Machines (SVMs), Lagrange multipliers are used to convert a constrained optimization problem into a form that can be solved more easily, specifically to find the optimal hyperplane that separates data points into different classes.

**Q 5 What are Support Vectors in SVM?**
  - In a Support Vector Machine (SVM), support vectors are the data points that are closest to the decision boundary (hyperplane) separating different classes

**Q 6 What is a Support Vector Classifier (SVC)?**
  - A Support Vector Classifier (SVC) is a specific type of Support Vector Machine (SVM) used for classification tasks

**Q 7 What is a Support Vector Regressor (SVR)?**
  - Support Vector Regression (SVR) is a machine learning algorithm used for regression tasks, which aims to predict continuous numerical values

**Q 8 What is the Kernel Trick in SVM?**
  - The kernel trick in Support Vector Machines (SVMs) is a technique that allows SVMs to classify data that is not linearly separable by implicitly mapping the data into a higher-dimensional feature space where a linear separator can be used

**Q 9 Compare Linear Kernel, Polynomial Kernel, and RBF Kernel?**
  - Linear, polynomial, and RBF kernels are used in Support Vector Machines (SVMs) to map data into a higher-dimensional space, enabling the algorithm to find optimal decision boundaries for classification or regression

**Q 10 What is the effect of the C parameter in SVM?**
  - In Support Vector Machines (SVMs), the C parameter controls the trade-off between achieving a low training error and a smooth decision boundary

**Q 11 What is the role of the Gamma parameter in RBF Kernel SVM?**
  - In the context of RBF (Radial Basis Function) kernel SVM, the gamma parameter (often denoted as γ) controls the influence of individual training samples on the decision boundary

**Q 12 What is the Naïve Bayes classifier, and why is it called "Naïve"?**
  - The Naive Bayes classifier is a probabilistic classification algorithm that uses Bayes' theorem with the "naive" assumption of feature independence.
  - It's called "naive" because it simplifies the calculation by assuming that features are conditionally independent of each other given the class label.

**Q 13 What is Bayes’ Theorem?**
  - Bayes' Theorem is a mathematical formula that describes the probability of an event based on prior knowledge of conditions related to the event

**Q 14 Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes?**
  - The key difference between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes lies in how they handle the type of data they are given: Gaussian NB assumes continuous data, Multinomial NB handles discrete counts, and Bernoulli NB deals with binary features.

**Q 15 When should you use Gaussian Naïve Bayes over other variants?**
  - Gaussian Naive Bayes is best used when dealing with continuous numerical data that follows a Gaussian (normal) distribution.

**Q 16 What are the key assumptions made by Naïve Bayes?**
  - Naive Bayes classifiers make a strong simplifying assumption that features are conditionally independent given the class label

**Q 17 What are the advantages and disadvantages of Naïve Bayes?**
  - Naive Bayes classifiers are known for their simplicity and speed, making them easy to implement and efficient, especially with large datasets. However, they rely on the strong assumption of feature independence, which is often unrealistic in real-world scenarios, potentially leading to lower accuracy.

**Q 18 Why is Naïve Bayes a good choice for text classification?**
  - Naive Bayes is a good choice for text classification primarily because it's simple, fast, and effective, especially with high-dimensional data like text

**Q 19 Compare SVM and Naïve Bayes for classification tasks?**
  - SVM (Support Vector Machine) and Naïve Bayes are both popular classification algorithms, but they differ significantly in their approach and performance characteristics. SVM is generally more accurate, particularly when dealing with complex relationships between features, but can be more computationally expensive. Naïve Bayes, on the other hand, is simpler, faster, and performs well when features are relatively independent

**Q 20 How does Laplace Smoothing help in Naïve Bayes?**
  - Laplace smoothing, also known as add-one smoothing, is a technique used in Naive Bayes to prevent probabilities from becoming zero

**Q 4 Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision
boundary?**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import SVC

X, y = make_blobs(n_samples=300, centers=2, random_state=6, cluster_std=3.0)

poly_svm = SVC(kernel='poly', degree=3, C=1.0)
poly_svm.fit(X, y)

h = .02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

Z = poly_svm.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
plt.title('SVM with Polynomial Kernel Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

**Q 7 Write a Python program to train an SVM Classifier with different C values and compare the decision
boundaries visually?**

```python
from sklearn.datasets import make_classification
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, random_state=0)
for C in [0.1, 1, 10, 100]:
    svm = SVC(kernel='linear', C=C)
    svm.fit(X, y)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    Z = svm.decision_function(xy).reshape(XX.shape)
    ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
    plt.title(f'SVM with C={C}')
    plt.show()
```

**Q 18 Write a Python program to train an SVM Classifier using Stratified K-Fold Cross-Validation and compute the
average accuracy?**

```python
from sklearn.datasets import load_wine
data=load_wine()
X=data.data
y=data.target
from sklearn.model_selection import StratifiedKFold
skf=StratifiedKFold(n_splits=5)
from sklearn.svm import SVC
model=SVC()
import numpy as np
from sklearn.model_selection import StratifiedKFold
accuracy_scores = []
for train_index, test_index in skf.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

average_accuracy = np.mean(accuracy_scores)
print(f"Average Accuracy with Stratified K-Fold Cross-Validation: {average_accuracy}")
```

**Q 20 Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and
compare accuracy?**

```python
import pandas as pd
from sklearn.datasets import load_wine
data=load_wine()
X=pd.DataFrame(data.data,columns=data.feature_names)
y=pd.Series(data.target)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
from sklearn.svm import SVC
from sklearn.feature_selection import RFE
estimator=SVC(kernel='linear')
selector=RFE(estimator,n_features_to_select=2,step=1)
from sklearn.pipeline import make_pipeline
model=make_pipeline(selector,SVC(kernel='linear'))
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy_with_rfe:",accuracy)
model=SVC(kernel='linear')
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy_without_rfe:",accuracy)
```

**Q 25 Write a Python program to train a Naïve Bayes classifier and evaluate its performance using the ROC-AUC
score?**

```python
from sklearn.datasets import load_iris
data=load_iris()
X=data.data
y=data.target
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(X_train,y_train)
y_pred_proba=model.predict_proba(X_test)
from sklearn.metrics import roc_auc_score
roc_auc=roc_auc_score(y_test,y_pred_proba,multi_class='ovr')
print("ROC-AUC Score:",roc_auc)
```

**Q 26 Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve?**




```python
from sklearn.datasets import load_iris
data=load_iris()
X=data.data
y=data.target
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
from sklearn.svm import SVC
model=SVC(probability=True)
model.fit(X_train,y_train)
y_pred_proba=model.predict_proba(X_test)
from sklearn.metrics import precision_recall_curve
precision,recall,thresholds=precision_recall_curve(y_test,y_pred_proba[:,1])
import matplotlib.pyplot as plt
plt.plot(recall,precision)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.show ()
```
